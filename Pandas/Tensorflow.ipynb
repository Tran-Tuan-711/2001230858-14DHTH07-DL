```python
import tensorflow as tf

# Bai 1
print('='*30 + 'Bai 1' + '='*30)
t = tf.zeros((2,3))
print(t)

# Bai 2
print('='*30 + 'Bai 2' + '='*30)
X = [[1,2,3],[4,5,6]]
t = tf.zeros_like(X)
print(t)

# Bai 3
print('='*30 + 'Bai 3' + '='*30)
t = tf.fill((3, 2), 5)
print(t)

# Bai 4
print('='*30 + 'Bai 4' + '='*30)
choices = tf.constant([6, 8, 10])
idx = tf.random.uniform((50,), minval=0, maxval=3, dtype=tf.int32)
t = tf.gather(choices, idx)
print(t)

# Bai 5
print('='*30 + 'Bai 5' + '='*30)
t = tf.range(10, 101, 2)
print(t)

# Bai 6
print('='*30 + 'Bai 6' + '='*30)
t = tf.random.normal((3,2), mean=0, stddev=2)
print(t)

# Bai 7
print('='*30 + 'Bai 7' + '='*30)
t = tf.random.uniform((3,2), minval=0, maxval=2)
print(t)

# Bai 8
print('='*30 + 'Bai 8' + '='*30)
X = tf.constant([[1,2],[3,4],[5,6]])
t = tf.random.shuffle(X)
print(t)

# Bai 9
print('='*30 + 'Bai 9' + '='*30)
t = tf.random.uniform((10,10,3))
crop = tf.image.random_crop(t, size=(5,5,3))
print(crop.shape)

# Bai 10
print('='*30 + 'Bai 10' + '='*30)
x = tf.random.uniform((5,), minval=0, maxval=10, dtype=tf.int32)
y = tf.random.uniform((5,), minval=0, maxval=10, dtype=tf.int32)
result = tf.where(x < y, x + y, x - y)
print('x:', x)
print('y:', y)
print('result:', result)

# Bai 11
print('='*30 + 'Bai 11' + '='*30)
x = tf.random.uniform((3,2), minval=5, maxval=10, dtype=tf.int32)
y = tf.random.uniform((2,3), minval=15, maxval=20, dtype=tf.int32)
diag = tf.linalg.diag(tf.linalg.diag_part(x))
result = diag + tf.transpose(y)
print(result)

# Bai 12
print('='*30 + 'Bai 12' + '='*30)
x = tf.random.uniform((3,2,2))
y = tf.random.uniform((3,2,2))
print(x + y)
print(x - y)
print(x * y)
print(x / y)

# Bai 13
print('='*30 + 'Bai 13' + '='*30)
class AddModel(tf.Module):
    def __call__(self, x, y):
        return x + y
model = AddModel()
print(model(tf.constant(2), tf.constant(3)))

# Bai 14
print('='*30 + 'Bai 14' + '='*30)
class MathModel(tf.Module):
    def __call__(self, x, y):
        return x - y, x * y
model = MathModel()
print(model(tf.constant(5), tf.constant(2)))

# Bai 15
print('='*30 + 'Bai 15' + '='*30)
class Rectangle(tf.Module):
    def __call__(self, a, b):
        return (a + b) * 2
model = Rectangle()
print(model(3,4))

# Bai 16
print('='*30 + 'Bai 16' + '='*30)
class Box(tf.Module):
    def __call__(self, a, b, c):
        Sxq = (a + b) * c * 2
        Stp = (a * b) * 2 + Sxq
        return Stp
model = Box()
print(model(2,3,4))

# Bai 17
print('='*30 + 'Bai 17' + '='*30)
def check_parallel(AD, BC, AB, CD):
    return (AD == BC) and (AB == CD)
print(check_parallel(5,5,4,4))

# Bai 18
print('='*30 + 'Bai 18' + '='*30)
def check_equal(AB, CD, AD, BC):
    return (AD == BC)
print(check_equal(4,4,5,5))

# Bai 19
print('='*30 + 'Bai 19' + '='*30)
def midline(AM, MB, AN, NC, BC):
    if AM == MB and AN == NC:
        return BC / 2
    return None
print(midline(2,2,3,3,10))
```
